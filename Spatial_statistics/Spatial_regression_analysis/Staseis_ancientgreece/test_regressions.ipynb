```python
import pandas as pd
from pysal.model import spreg
```

```python
db = pd.read_csv("polis data_distributed.csv")
```

```python
db = db.drop(['prom 2', 'prom 3','area 2', 'Bronze', 'Silver', 'Grid', 'Koinon', 'Pleiades link', 'Unnamed: 27'], axis=1)
```

```python
db = db.dropna()
```

```python
variable_names = ['Polisity', 'Hellenicity', 'In/out',
       'prom 1', 'area 1', 'Colonies', 'Victors', 'Proxenoi', 'Democracy',
       'Walls', 'Delian L', 'Region #', 'Latitude', 'Longitude',
       'Elevation m']
```

```python
# Fit OLS model
m1 = spreg.OLS(
    # Dependent variable
    db[["staseis"]].values,
    # Independent variables
    db[variable_names].values,
    # Dependent variable name
    name_y="staseis",
    # Independent variable name
    name_x=variable_names,
)
```

```python
print(m1.summary)
```
